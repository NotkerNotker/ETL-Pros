```python
import csv
import pandas as pd
from sqlalchemy import create_engine
from datetime import datetime
```

```python
# Read the Player Data File (CSV)
player_data_csv = "player_data.csv"
player_data = pd.read_csv(player_data_csv)
player_data.head()
```

```python
# Read the Players File (CSV)
player_csv = "Players.csv"
player = pd.read_csv(player_csv)
player.head()
```

```python
# Read the Player Stats File (CSV)
season_stats_csv = "Seasons_Stats.csv"
season_stats = pd.read_csv(season_stats_csv)
season_stats.head()
```

```python
#Clean up: Get only the players who played from 1980 on
player_data_year = player_data.loc[player_data['year_start'] >= 1980]
```

```python
#Clean up: Get only the player stats who played from 1980 on
season_stats_year = season_stats.loc[season_stats['Year'] >= 1980]
season_stats_year.head()
```

```python
players_df = pd.DataFrame(player_data_year['name'])
#players_df['first_name'] = player_data['name'].str.split(' ',expand=True)
#players_df['last_name']
players_df.head()
```

```python
new = players_df["name"].str.split(" ", n = 1, expand = True)
players_df['first_name'] = new[0]
players_df['last_name'] = new[1]
players_df.rename(columns = {'name':'player_name'}, inplace = True)
players_df.head()
```

```python
players_df= players_df.dropna()
players_df.describe()
```

```python
rds_connection_string = "postgresql://postgres:rootpassword@localhost:5432/ETL_Work"
engine = create_engine(rds_connection_string)
engine.table_names()
```

```python
players_df.to_sql(name='players', con=engine, if_exists='append', index=False)
```

```python
player_lookup = pd.DataFrame(pd.read_sql_query('select * from players', con=engine))
```

```python
player_lookup.head()
```

```python
player_lookup.to_csv("Player_lookup.csv", index = False)
```

```python
Player_NBA = pd.DataFrame(player_data_year[['name', 'year_start', 'year_end', 'height', 'weight']])
Player_NBA.rename(columns = {'name':'player_name'}, inplace = True)
```

```python
Player_NBA.head()
```

```python
# Merge two dataframes using an outer join
Player_NBA_id = pd.merge(Player_NBA, player_lookup, on="player_name")
Player_NBA_id.head()
```

```python
players_NBA_df= Player_NBA_id.dropna()
players_NBA_df.head()
```

```python
Player_NBA_Information = pd.DataFrame(players_NBA_df[['player_id', 'year_start', 'year_end', 'height', 'weight']])
Player_NBA_Information.dtypes
```

```python
Player_NBA_Information.to_sql(name='player_nba_information', con=engine, if_exists='append', index=False)
```

```python
player_nba_info = pd.DataFrame(pd.read_sql_query('select * from Player_NBA_Information', con=engine))
player_nba_info.head()
```

```python
player_nba_info.to_csv("Player_NBA_Information.csv", index = False)
```

```python
player_nba_stats = pd.DataFrame(NBA_stats[['player_id', 'Age', 'Year', 'Tm','Pos', 'MP', 
                                           'PTS', '2P', '3P', 'FTr', 'AST',
                                          'STL', 'BLK', 'TOV', 'FG%']])
player_nba_stats.rename(columns = {'Age':'age',
                                   'Year':'year',
                                   'Tm':'team',
                                   'Pos':'position_played',
                                   'MP':'minutes_played',
                                   'PTS':'points_scored',
                                   '2P':'two_points_scored',
                                   '3P':'three_points_scored',
                                   'FTr':'free_throws_made',
                                   'AST':'assists',
                                   'STL':'steals',
                                   'BLK':'blocks',
                                   'TOV':'turnovers',
                                   'FG%':'field_goal_percentage'}, inplace = True)
player_nba_stats.dtypes
```
